```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# Define stock ticker for Reliance (NSE)
ticker = "RELIANCE.NS"

# Download last 1 year of stock data
stock_data = yf.download(ticker, period="1y")

# Reset index to make Date a column
stock_data.reset_index(inplace=True)

# View first 5 rows
stock_data.head()
```

```python
# Check total number of rows
stock_data.shape
```

```python
# Save raw stock data to raw data folder
stock_data.to_csv("../data/raw/reliance_stock_raw.csv", index=False)

print("Raw stock data saved successfully.")
```

```python
# Make a copy of raw data for cleaning
clean_stock = stock_data.copy()

# Ensure Date column is datetime
clean_stock['Date'] = pd.to_datetime(clean_stock['Date'])

# Sort data by Date (ascending)
clean_stock = clean_stock.sort_values('Date')

# Check for missing values
clean_stock.isnull().sum()
```

```python
# Save cleaned stock data to processed layer
clean_stock.to_csv("../data/processed/reliance_stock_cleaned.csv", index=False)

print("Cleaned stock data saved successfully.")
```
